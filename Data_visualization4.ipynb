```
#importing libraries
import requests
```

```
from bs4 import BeautifulSoup
```

```
from collections import Counter
```

```
import matplotlib.pyplot as plt
```

```
# to get data from given url
webdata=requests.get( 'https://en.wikipedia.org/wiki/Machine_learning')
```

```
webdata    # will give response only
```

```
htmldata=webdata.text
```

```
soup=BeautifulSoup(htmldata,'html.parser')   # parser 
```

```
# cleaning that data for readable form
clean_data=soup.get_text()
```

```
clean_data=clean_data.replace('\n',' ')
```

```
clean_data=clean_data.replace('.',' ')
#clean_data
```

```
# to count all the words after scraping
count=0
list2=clean_data.split(" ")
for words in list2:
  count=count+1
print(count)
```

```
# removing spaces from the list
while ('' in list2):
  list2.remove('')
```

```
# creating dictionary for representation of repeation of words
ls=[]
for i in list2:
  word_count = list2.count(i)   # Pythons count function 
  ls.append((i,word_count))       


diction = dict(ls)
#print(diction)
```

```
# creation of dictionary of top20 words
list1=[]
k=Counter(diction)
high=k.most_common(20)
for i in high:
  list1.append((i[0],i[1]))
list1=dict(list1)
print(list1)
```

```
# pie chart of top 20 repeated words
plt.figure(figsize=(5,8))
plt.pie(list1.values(),labels=list(list1.keys()),autopct='%1.1f%%')
plt.show()
          
```

```
# bar graph of top 20 repeated words
plt.figure(figsize=(12,5))
plt.bar(range(len(list1)), list1.values(),width=0.2)
plt.xticks(range(len(list1)), list(list1.keys()))
plt.show()
```
