```
from sklearn.datasets import load_digits 
```

```
# loading Digit data
digit=load_digits()
```

```
# desc
dir(digit)
```

```
# features
features=digit.data
```

```
# Label
label=digit.target
```

```
# training and testing
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label=train_test_split(features,label,test_size=0.2)
```

```
import matplotlib.pyplot as plt
plt.imshow(digit.images[0])
```

```
# now calling SVC
from sklearn.svm import SVC
```

```
# calling clf (read gamma and kernel argument of SVC)
clf = SVC()
```

```
# now training my classifier
trained=clf.fit(train_data,train_label)
```

```
#now time for prediction
output=trained.predict(test_data)
output
```

```
test_label
```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(output,test_label)
```
