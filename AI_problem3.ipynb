```
# importing libraries
import pandas as pd
```

```
import matplotlib.pyplot as plt
```

```
# reading data from URL
df =pd.read_csv("http://13.234.66.67/summer19/datasets/bank.csv")
```

```
df.info()
```

```
df
```

```
# plotting pie chart for no. of customers living in France,Germany,Spain
plt.pie(df['Geography'].value_counts().values,labels=df['Geography'].value_counts().index,autopct='%1.1f%%')
```

```
# no. of male and female customer
a=df['Gender'].value_counts().values
dg=pd.DataFrame({'Gender':['Male','Female'],'count': [a[0],a[1]]})
ax = dg.plot.bar(x='Gender',y='count',rot=1000,width=0.2)
```

```
# avg salary of male and female customer
sal=df.groupby('Gender')['EstimatedSalary'].mean().values
dg=pd.DataFrame({'Count':a,'avg_salary':sal},index=['Male','Female'])
ax = dg.plot.bar(width=0.3)
```

```
#  average age of male and female
age=df.groupby('Gender')['Age'].mean().values
dg=pd.DataFrame({'avg_age':age},index=['Male','Female'])
ax = dg.plot.bar(rot=0,width=0.2)
```

```
# histogram for credit score of customer
df[['CreditScore']].plot(kind='hist',bins=[300,400,500,600,700,800,900],rwidth=0.8,color='green')
```

```
# pie chart for active and inactive member
plt.pie(df['IsActiveMember'].value_counts().values,labels=['Active Member','Inactive Member'],autopct='%1.1f%%')
```

```
# histogram for tenure of customers
df[['Tenure']].plot(kind='hist',bins=[0,2,4,6,8,10],rwidth=0.8,color='pink')
```
